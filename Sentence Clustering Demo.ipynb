```python
# Import data from Excel workbook.
# There should be one sheet (Sheet1) with headings id & description.
# This is the list of sentences you want to use to create clusters.

import pandas as pd

DataFile = "C:\\Users\\u0396993\\Documents\\junk\\SentenceListForClustering.xlsx"

df1 = pd.read_excel(DataFile, 'Sheet1')

# Add the ids and text chunks to lists.
Sheet1_id = df1.id.values.tolist()
Sheet1_text = df1.description.values.tolist()
```

```python
from sentence_transformers import SentenceTransformer

# Models - https://huggingface.co/models?library=sentence-transformers
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
embeddings1 = model.encode(Sheet1_text)
```

```python
# clustering https://www.youtube.com/watch?v=OlhNZg4gOvA time index 22:00

from sklearn.cluster import KMeans
import numpy as np

# normalize the embeddings to unit length
embeddings_norm = embeddings1 / np.linalg.norm(embeddings1, axis=1, keepdims=True)
```

```python
clustering_model = KMeans(n_clusters=60)
clustering_model.fit(embeddings_norm)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```

```python
clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []
        
    clustered_sentences[cluster_id].append(Sheet1_text[sentence_id])
clustered_sentences
```
